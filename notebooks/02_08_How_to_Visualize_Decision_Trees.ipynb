## Import Libraries

```python
%matplotlib inline

import matplotlib.pyplot as plt
import pandas as pd

from sklearn.datasets import load_iris
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier

from sklearn import tree
```

## Load the Dataset
The Iris dataset is one of datasets scikit-learn comes with that do not require the downloading of any file from some external website. The code below loads the iris dataset.

```python
# Load the Iris dataset
iris = load_iris()

# Split the dataset into features (X) and target (y)
X = iris.data
y = iris.target
df.head()
```

## Split Data into Training and Test Sets

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
```

## Scikit-learn 4-Step Modeling Pattern

### Step 1: Import the model you want to use

In sklearn, all machine learning models are implemented as Python classes

```python
# This was already imported earlier in the notebook so commenting out
#from sklearn.tree import DecisionTreeClassifier
```

### Step 2: Make an instance of the Model

```python
# Create an instance of the Decision Tree model
clf = DecisionTreeClassifier(random_state=42)
```

### Step 3: Training the model on the data, storing the information learned from the data

Model is learning the relationship between x (features: sepal width, sepal height etc) and y (labels-which species of iris)

```python
# Train the model on the training data
clf.fit(X_train, y_train)
```

### Step 4: Optional: Make predictions on the test set

Uses the information the model learned during the model training process

```python
# Predict for One Observation (image)
y_pred = clf.predict(X_test)
```

Predict for Multiple Observations (images) at Once

```python
clf.predict(X_test[0:10])
```

## Measuring Model Performance

```python
plt.figure(figsize=(20, 10))  # Adjusts the figure size for better readability
tree.plot_tree(clf, filled=True)
plt.show()
```

Accuracy is defined as:
(fraction of correct predictions): correct predictions / total number of data points

```python
score = clf.score(X_test, y_test)
print(score)
```

## How to Visualize Decision Trees using Matplotlib

#### Default Visualization Based on the Model

```python
# Enhanced visualization of the decision tree
plt.figure(figsize=(20, 10))  # Adjust the size of the plot for better readability
tree.plot_tree(
    clf,
    feature_names=iris.feature_names,  # Add feature names for better context
    class_names=iris.target_names,     # Add class names for interpretation
    filled=True,                       # Color the nodes based on the majority class
    rounded=True                       # Use rounded edges for improved aesthetics
)
plt.show()
```

```python
# Adjust the figure size and DPI for better readability
plt.figure(figsize=(20, 10), dpi=100)  # Set a larger size and higher DPI for clarity
tree.plot_tree(
    clf,
    feature_names=iris.feature_names,  # Add feature names for better context
    class_names=iris.target_names,     # Add class names for interpretation
    filled=True,                       # Color the nodes based on the majority class
    rounded=True                       # Use rounded edges for improved aesthetics
)
plt.show()
```
